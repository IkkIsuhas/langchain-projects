```python
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from langchain_community.document_loaders import WikipediaLoader
from langgraph.graph import StateGraph,END
from dotenv import load_dotenv
import wikipedia
import os
import getpass
load_dotenv()
tavily_api_key= os.getenv("TAVILY_API_KEY")
```

```python
def wiki(state):
    print("Loading Wikipedia tool...")
    wikipedia.set_user_agent("ReAct-Agents")
    wiki_tool = WikipediaLoader(
        query=state["answer"],
        load_max_docs=2
    )
    result = wiki_tool.load()
    return {
        "answer":result
    }

def web_tool(state):
    print("Loading web-search tool...")
    query = state["question"]
    
    search_tool = TavilySearch(
            max_results=1
        )
    result = search_tool.invoke(query)
    return {
            "answer" : result
        }
```

```python
web_result = web_tool({
    "question":'what is langchain'
})
print(web_result["answer"]["results"][0]["content"])
```

```python
def llm_model():
    llm = ChatOpenAI(
        model = 'openai/gpt-oss-120b:free',
        api_key=os.getenv('OPENROUTER_API_KEY'),
        base_url="https://openrouter.ai/api/v1"
    )
    return llm
```

```python
llm = llm_model()
result = llm.invoke('who are you?')
print(result.content)
```
